```python
import requests
import csv
import time

# ==============================
# Configuration
# ==============================

TOKEN = "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"   # <-- NEVER hardcode real tokens in public code
HEADERS = {"Authorization": f"token {TOKEN}"}

search_terms = [

# Audit Trail tactic (extended)
"audit trail",
"audit log",
"audit logging",
"auditing",
"activity log",
"event log",
"event logging",
"transaction log",
"access log",
"security log",
"user activity tracking",
"user activity log",
"traceability",
"trace log",
"change log",
"changelog",
"operation log",
"history tracking",
"data lineage",
"accountability",
"non-repudiation",
"tamper-proof log",
"immutable log",
"append-only log",
"log integrity",
"log retention",
"log compliance",
"compliance logging",
"forensic logging",
"forensics",
"monitoring log",
"request tracing",
"distributed tracing",
"audit record",
"security auditing",
"regulatory compliance",
"gdpr logging",
"hipaa compliance",

# Resource Pooling tactic (extended)
"resource pooling",
"resource pool",
"object pool",
"object pooling",
"connection pool",
"database connection pool",
"thread pool",
"worker pool",
"process pool",
"pool manager",
"pool size",
"pool exhaustion",
"pool configuration",
"pooled resource",
"resource reuse",
"resource reuse strategy",
"instance pooling",
"session pool",
"http connection pool",
"socket pool",
"buffer pool",
"memory pool",
"pool allocation",
"bounded pool",
"fixed size pool",
"dynamic pool",
"pool initialization",
"pool shutdown",
"borrow return pattern",
"acquire release pattern",

# Scheduling tactic (extended)
"scheduling",
"scheduler",
"task scheduling",
"job scheduling",
"process scheduling",
"thread scheduling",
"priority scheduling",
"priority queue",
"task queue",
"job queue",
"work queue",
"queueing",
"fair scheduling",
"round robin",
"round-robin",
"fifo scheduling",
"deadline scheduling",
"rate monotonic",
"earliest deadline first",
"edf",
"cron",
"cron job",
"timer",
"timed task",
"periodic task",
"delayed task",
"task dispatcher",
"dispatcher",
"orchestration",
"workflow engine",
"workflow scheduling",
"resource scheduling",
"cpu scheduling",
"load shedding",
"admission control",
"quota",
"resource allocation",
"resource arbitration",

# Core performance concepts
"slowness",
"bottleneck",
"overhead",

# Resource utilization
"cpu usage",
"memory usage",
"memory leak",
"heap",
"gc",
"garbage collection",
"resource consumption",

# Optimization tactics
"optimize",
"optimization",
"efficient",
"efficiency",
"speed up",

# Caching tactics
"cache",
"caching",
"memoization",

# Concurrency & parallelism
"thread",
"multithreading",
"concurrency",
"parallel",
"asynchronous",
"async",
"non-blocking",
"blocking",

# Load handling
"scalability",
"scalable",
"load balancing",
"high load",
"traffic spike",
"burst",

# I/O performance
"io",
"disk",
"bandwidth",
"serialization",
"deserialization",

# Database performance
"query optimization",
"slow query",
"index",
"database performance",

# Architecture-level tactics
"lazy loading",
"batch processing",
"pooling",
"connection pool",
"thread pool",
"buffer",
"buffering",
"streaming",
"pagination",

# Monitoring / detection
"profiling",
"profiler",
"benchmark",
"metrics",
"monitoring",
"tracing",

# Fault-related performance issues
"timeout",
"retry",
"backoff",
"throttle",
"rate limit",

# Event-based / heartbeat (extended)
"heartbeat",
"healthcheck",
"health check",
"liveness",
"liveness probe",
"readiness",
"readiness probe",
"keepalive",
"keep-alive",
"ping",
"pong",
"watchdog",
"watchdog timer",
"failure detection",
"failure detector",
"node failure",
"node health",
"cluster health",
"service health",
"crash detection",
"replica health",
"replication lag",
"leader election",
"consensus",
"raft",
"gossip protocol",
"cluster membership",
"service discovery",
"service registry",
"high availability",
"failover",
"auto recovery",
"self-healing"

]



BASE_FILTER = "language:Python stars:>=50"
PER_PAGE = 100

# ==============================
# Helper Function
# ==============================

def search_github(term):
    page = 1
    collected = []

    while True:
        query = f'"{term}" {BASE_FILTER}'
        url = (
            f"https://api.github.com/search/repositories"
            f"?q={query}&page={page}&per_page={PER_PAGE}"
            f"&sort=stars&order=desc"
        )

        response = requests.get(url, headers=HEADERS)

        if response.status_code != 200:
            print(f"Failed for term '{term}': {response.status_code}")
            break

        data = response.json()
        items = data.get("items", [])

        if not items:
            break

        collected.extend(items)
        page += 1

        # Prevent rate limiting
        time.sleep(1)

    return collected


# ==============================
# Main Execution
# ==============================

all_repos = {}
total_queries = 0

for term in search_terms:
    print(f"Searching for: {term}")
    results = search_github(term)
    total_queries += 1

    for repo in results:
        # Deduplicate using repository ID
        all_repos[repo["id"]] = repo

print(f"\nTotal unique repositories found: {len(all_repos)}")

# ==============================
# Save to CSV
# ==============================

csv_filename = "Python_github_repositories.csv"

with open(csv_filename, mode="w", newline="", encoding="utf-8") as file:
    writer = csv.DictWriter(
        file,
        fieldnames=["name", "url", "description", "stars", "language", "owner"]
    )

    writer.writeheader()

    for repo in all_repos.values():
        writer.writerow({
            "name": repo["name"],
            "url": repo["html_url"],
            "description": repo.get("description", "No description"),
            "stars": repo["stargazers_count"],
            "language": repo.get("language", "N/A"),
            "owner": repo["owner"]["login"]
        })

print(f"Results saved to {csv_filename}")
```
